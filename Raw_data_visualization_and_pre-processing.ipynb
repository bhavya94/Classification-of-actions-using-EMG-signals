### For the purpose of visualization, we will check each channel of each subject. Since there are 4 subjects performing 20 actions and each action has 8 channels, we have 640 possible visualisations. However, the purpose is to understand the difference between the subjects and standardize the data for modelling. Hence, we take standing as our basis for comparision in 'normal' activities since we do not expect high involvement of muscles, making it easy for us to normalize between subjects. We will also visualize pushing from 'aggressive' actitivies since we believe it is one of the actions that involes most of the muscles in the body.

```python
def raw_visualize(data):
    index = data.index.values
    for i in range(0,len(data.columns)):
        plt.subplots(1)
        plt.title(data.columns[i])
        channel = list(data.loc[:,data.columns[i]])
        plt.plot(index,channel)
        plt.xlabel("Hz")
        plt.ylabel("mV")
        plt.show
    
```

## The data, however, is noisy and unclean. There are two issues when collecting signals from electrodes attached to skin surface: signal-to-noise ratio and distortion of signals. Noise is added to the data since the electrical signal interacts with blood vessels and tissues between the skin and muscle. To improve the signal-to-noise ratio, we detrended the raw data to filter out the DC offset. Next, we rectified the data by converting the signal to a single polarity, to ensure that signals don’t average to 0 during analysis. The Savitzky-Golay filter, set to an order of 5 and framelength of 49, was then applied to smoothen the data and further eliminate noise.
